```python
tempdata1 = pd.Series(raw_data).to_frame().reset_index().rename(columns={0:'value', 'index':'relation'})
tempdata1['single'] = tempdata1['relation'].apply(lambda x: x.split('&'))
tempdata1
```

```python
# tempdata1.drop(labels=['relation'], axis=1)#.explode(column='single')
base_data1 = tempdata1.explode(column='single').pivot_table(index='relation', values='value', columns='single', fill_value=0)
base_data1
```

```python
# p1, p2 = np.sort(base_data1.sum(axis=1), )
value_upper = base_data1.max(axis=1).values

base_data2 = base_data1.iloc[value_upper.argsort()[::-1], :]
value_horizon = base_data2.sum(axis=0).values
base_data3 = base_data2.iloc[:, value_horizon.argsort()].T
base_data3.columns = range(base_data3.columns.shape[0])
base_data3
```

```python
# base_data3.index

# plot_data4 = base_data3.values

fill_color = "#242123"
empty_color = "#C4BEC2"

max_show = 40

plot_data4 = base_data3.loc[:, :max_show]

plot_data41 = plot_data4.reset_index().melt(id_vars='single', value_name='value', var_name='numid')
def value2_color(value):
    if value > 0:
        return fill_color
    else:
        return empty_color
    
plot_data41['color'] = plot_data41['value'].apply(lambda x: value2_color(x))
# plot_data41
```
